```python
from apyori import apriori
```

```python
data = pd.read_csv("store_data.csv")
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.fillna(0, inplace = True)
```

```python
data.head()
```

```python
# records = []
# for i in range(0, len(data)):
#     records.append([str(data.values[i, j])for j in range(0,20) if str(data.values[i, j] != '0')])
records = data.replace('0', pd.NA).applymap(str).stack().groupby(level=0).apply(list).tolist()
```

```python
records[0]
```

```python
records[1]
```

```python
association_rules = apriori(records, min_support = 0.004, min_confidence = 0.2, min_lift = 2, min_length = 2)
association_rules
```

```python
result = list(association_rules)
result
```

```python
print("There are {} relation derived".format(len(result)))
```

```python
for i in range(0, len(result)):
    print(result[i][0])
```

```python
for item in result:
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + "-" + items[1])
    
    print("Support: "+str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
    
```
